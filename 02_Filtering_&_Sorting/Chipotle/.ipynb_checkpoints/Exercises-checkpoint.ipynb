### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep = '\t')
```

```python
chipo.dtypes
```

```python
chipo.head()
```

```python
chipo = chipo.assign(
    item_price = lambda x : x.item_price.str.replace("$","").astype(float)
)
```

### Step 4. How many products cost more than $10.00?

```python
#I answered the wrong question
(chipo.item_price > 10)
```

```python
(chipo.item_price > 10).sum()
```

```python
# to answer this question
# assume a "product" is unique combinations of quantity, item_name and choice_description
unique_prod = chipo.drop_duplicates( ['quantity', 'item_name', 'item_price'])
unique_prod.head()
```

```python
# count the number of uniqie products with item_price > $10
unique_prod.query('item_price > 10').count()
```

### Step 5. What is the price of each item?

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# How many orders contained a Veggie Salad Bowl
chipo.item_name.str.contains('Veggie Salad Bowl').sum()
```

```python
# How many times was a Veggie Salad Bowl ordered
chipo[chipo.item_name.str.contains('Veggie Salad Bowl')]['quantity'].sum()
```

```python
chipo.groupby('item_name').quantity.sum()['Veggie Salad Bowl']
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo[ (chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)   ]
```

```python
chipo[ (chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)   ].shape[0]
```
