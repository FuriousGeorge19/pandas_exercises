### Step 8. Group the year by decades and sum the values

```python
# I want population to be aggregated by mean and the rest by sum without
# manually assigning each column in agg
```

```python
# look at columns
crime.columns
```

```python
# create an aggregate dictionary; Population is mean, everything else is sum
agg_dict = { x : (np.sum if x != 'Population' else np.mean) for x in crime.columns   }
agg_dict
```

```python
# apply `agg_dict` created above
crime_10y = crime.groupby(pd.Grouper(freq='10Y')).agg(agg_dict)
crime_10y
```

```python
# THE SOLUTION THEY USE WITH resampling is better than mine with pd.Grouper, because it is more flexible
# as it allows you to choose whether or not to start from the beginning of the initial year or the end
crime_10y_v2 = crime.resample('10AS').agg(agg_dict)
crime_10y_v2
```

### Step 9. What is the most dangerous decade to live in the US?

```python
# I'll use Violent as a Percent of Population
crime_10y.assign(
    violent_pct = lambda x: (x.Violent.div(x.Population)*100).round(2)
) \
.sort_values(by='violent_pct', ascending=False)
```
